```python
import cirq

# Define los qubits
q0 = cirq.LineQubit(0)
q1 = cirq.LineQubit(1)
q2 = cirq.LineQubit(2)

# Define los bits clásicos
c0 = cirq.LineQubit(3)
c1 = cirq.LineQubit(4)

# Define el circuito cuántico
circuito = cirq.Circuit()

# Aplica las compuertas cuánticas
circuito.append(cirq.H(q0))  # Compuerta Hadamard en el qubit q0
circuito.append(cirq.CNOT(q0, q1))  # Compuerta CNOT entre q0 y q1
circuito.append(cirq.CNOT(q1, q2))  # Compuerta CNOT entre q1 y q2

# Mide los qubits en los bits clásicos
circuito.append(cirq.measure(q0, c1))  # Mide q0 y almacena el resultado en el bit clásico c0
circuito.append(cirq.measure(q2, c0))  # Mide q1 y almacena el resultado en el bit clásico c1

print("Circuito cuántico:")
print(circuito)
```

```python
#Ejemplo 3
import cirq
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir el qubit
qubit = cirq.LineQubit(0)

# Definir la compuerta H y aplicarla al qubit
h_gate = cirq.H(qubit)

# Crear el circuito con la compuerta H
circuit = cirq.Circuit(h_gate)

# Simular el circuito en un simulador de Cirq
simulator = cirq.Simulator()
result = simulator.simulate(circuit)

# Obtener el estado final del qubit
final_state = result.final_state_vector

# Calcular los ángulos de la esfera de Bloch
x = 2 * np.real(final_state[0] * np.conj(final_state[1]))
y = 2 * np.imag(final_state[0] * np.conj(final_state[1]))
z = np.abs(final_state[0])**2 - np.abs(final_state[1])**2

# Visualizar el estado en la esfera de Bloch
bloch_sphere = plt.figure()
ax = bloch_sphere.add_subplot(111, projection='3d')

ax.plot([0, x], [0, y], [0, z], marker='o')
ax.text(x, y, z, '|ψ⟩')

ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
#Ejemplo 5
import cirq

# Definir los qubits
qubits = cirq.LineQubit.range(2)

# Crear el circuito cuántico
circuit = cirq.Circuit()

# Aplicar la compuerta H (Hadamard) a cada qubit
for qubit in qubits:
    circuit.append(cirq.H(qubit))

# Visualizar el circuito
print("Circuito cuántico:")
print(circuit)

# Simular el circuito
simulator = cirq.Simulator()
result = simulator.simulate(circuit)

# Imprimir el estado final
print("\nEstado final:")
print(result.final_state_vector)
```
